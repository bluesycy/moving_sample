```python
import numpy as np
import matplotlib.pyplot as plt

from skimage import data
from skimage.registration import phase_cross_correlation
from skimage.registration._phase_cross_correlation import _upsampled_dft
from scipy.ndimage import fourier_shift
import scipy.ndimage as ndimage
from tifffile import tifffile
```

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt
from IPython.display import clear_output
from time import sleep
```

```python
data_path = "/nfs/data8/chuyu/data/20250301_183730/20250301_183730_confocal.tif"

images = tifffile.imread(data_path)
n_z = images.shape[0]
n_c = images.shape[1]
```

```python
which_channel = 1
volume = images[:,which_channel,:,:]
vmax = np.max(volume)
```

```python
for i in range(n_z):
    plt.figure(dpi=100)

    plt.imshow(volume[i, :, :], vmin = 0, vmax = vmax)
    plt.title(i)
    plt.colorbar()

    plt.show()
    clear_output(wait=True)
```

```python
image = volume[174,:,:]
offset_image = volume[175,:,:]
```

```python
# pixel precision first
shift, error, diffphase = phase_cross_correlation(image, offset_image)

fig = plt.figure(figsize=(8, 3))
ax1 = plt.subplot(1, 3, 1)
ax2 = plt.subplot(1, 3, 2, sharex=ax1, sharey=ax1)
ax3 = plt.subplot(1, 3, 3)

ax1.imshow(image, cmap='gray')
ax1.set_axis_off()
ax1.set_title('Reference image')

ax2.imshow(offset_image.real, cmap='gray')
ax2.set_axis_off()
ax2.set_title('Offset image')

# Show the output of a cross-correlation to show what the algorithm is
# doing behind the scenes
image_product = np.fft.fft2(image) * np.fft.fft2(offset_image).conj()
cc_image = np.fft.fftshift(np.fft.ifft2(image_product))
ax3.imshow(cc_image.real)
ax3.set_axis_off()
ax3.set_title("Cross-correlation")

plt.show()

print(f'Detected pixel offset (y, x): {shift}')
```

```python
# subpixel precision
shift, error, diffphase = phase_cross_correlation(
    image, offset_image, upsample_factor=100
)

fig = plt.figure(figsize=(8, 3))
ax1 = plt.subplot(1, 3, 1)
ax2 = plt.subplot(1, 3, 2, sharex=ax1, sharey=ax1)
ax3 = plt.subplot(1, 3, 3)

ax1.imshow(image, cmap='gray')
ax1.set_axis_off()
ax1.set_title('Reference image')

ax2.imshow(offset_image.real, cmap='gray')
ax2.set_axis_off()
ax2.set_title('Offset image')

# Calculate the upsampled DFT, again to show what the algorithm is doing
# behind the scenes.  Constants correspond to calculated values in routine.
# See source code for details.
cc_image = _upsampled_dft(image_product, 150, 100, (shift * 100) + 75).conj()
ax3.imshow(cc_image.real)
ax3.set_axis_off()
ax3.set_title("Supersampled XC sub-area")


plt.show()

print(f'Detected subpixel offset (y, x): {shift}')
```

```python
corrected_image = fourier_shift(np.fft.fftn(offset_image), shift)
corrected_image = np.fft.ifftn(corrected_image)
```

```python
image = volume[174,:,:]
offset_image = corrected_image
```

```python
# subpixel precision
shift, error, diffphase = phase_cross_correlation(
    image, offset_image, upsample_factor=100
)

fig = plt.figure(figsize=(8, 3))
ax1 = plt.subplot(1, 3, 1)
ax2 = plt.subplot(1, 3, 2, sharex=ax1, sharey=ax1)
ax3 = plt.subplot(1, 3, 3)

ax1.imshow(image, cmap='gray')
ax1.set_axis_off()
ax1.set_title('Reference image')

ax2.imshow(offset_image.real, cmap='gray')
ax2.set_axis_off()
ax2.set_title('Offset image')

# Calculate the upsampled DFT, again to show what the algorithm is doing
# behind the scenes.  Constants correspond to calculated values in routine.
# See source code for details.
cc_image = _upsampled_dft(image_product, 150, 100, (shift * 100) + 75).conj()
ax3.imshow(cc_image.real)
ax3.set_axis_off()
ax3.set_title("Supersampled XC sub-area")


plt.show()

print(f'Detected subpixel offset (y, x): {shift}')
```

```python
accumulated_shift = np.array([0.0, 0.0])
images_corrected = np.copy(images)

which_channel = 1
ref_volume = images[:,which_channel,:,:]

for i in range(n_z-1):
    image = ref_volume[i,:,:]
    offset_image = ref_volume[i+1,:,:]
    
    # subpixel precision
    shift, error, diffphase = phase_cross_correlation(
        image, offset_image, upsample_factor=100
    )
    
    
    print(f'Detected subpixel offset (y, x): {shift}')
    accumulated_shift += shift

    for i_channel in range(n_c):
        offset_image = images[i+1,i_channel,:,:]
        # corrected_image = fourier_shift(np.fft.fftn(offset_image), accumulated_shift)
        # corrected_image = np.fft.ifftn(corrected_image)
        corrected_image = ndimage.shift(offset_image, accumulated_shift)
        images_corrected[i+1,i_channel,:,:] = corrected_image
```
