```python
import nrrd
import skimage.measure as measure
import numpy as np
import scipy.ndimage as ndimage
from tqdm import trange

import torch
import torch.nn.functional as F
import torch.optim.lr_scheduler as lr_scheduler
```

```python
images, header = nrrd.read("20230307_121819round1_moving_sample.nrrd")
volume = images[:,:,::4]

volume = measure.block_reduce(volume, block_size=(2, 2, 1), func=np.mean)
```

```python
def torch_volume(data):
    volume = torch.tensor(data, dtype=torch.float32)  # Convert to PyTorch tensor
    return volume.unsqueeze(0).unsqueeze(0)  # Add batch and channel dimensions

class Translation2D(torch.nn.Module):
    def __init__(self):
        super(Translation2D, self).__init__()

        self.tx = torch.nn.Parameter(torch.tensor(0.0, requires_grad=True))       # Translation in x
        self.ty = torch.nn.Parameter(torch.tensor(0.0, requires_grad=True))       # Translation in y

    def forward(self, x):

        tx = self.tx / (x.shape[2] / 2)  # Normalize by width
        ty = self.ty / (x.shape[3] / 2)  # Normalize by height

        # Construct translation matrix
        transform_matrix = torch.eye(3, device=x.device)
        transform_matrix[0, 2] = ty
        transform_matrix[1, 2] = tx

        transform_matrix = transform_matrix[:2, :].unsqueeze(0)  # Add batch dimension and take 3x4 submatrix
        grid = F.affine_grid(transform_matrix, x.size(), align_corners=False)
        return F.grid_sample(x, grid, align_corners=False, mode="bicubic", padding_mode = "border")


def similarity_loss(fixed, moving_transformed):
    loss = F.mse_loss(fixed, moving_transformed)

    # loss.requires_grad = True
    return loss
def ncc_loss(fixed, moving_transformed):
    fixed_mean = torch.mean(fixed)
    moving_mean = torch.mean(moving_transformed)
    
    fixed_centered = fixed - fixed_mean
    moving_centered = moving_transformed - moving_mean
    
    numerator = torch.sum(fixed_centered * moving_centered)
    denominator = torch.sqrt(torch.sum(fixed_centered ** 2) * torch.sum(moving_centered ** 2))
    
    ncc = numerator / denominator
    return -ncc  # We return the negative NCC because we want to maximize NCC
```

```python
def register_2d(fixed_volume, moving_volume):

    model = Translation2D()
    optimizer = torch.optim.Adam(model.parameters(), lr= 0.001)  # Increased learning rate
    # scheduler = lr_scheduler.StepLR(optimizer, step_size=500, gamma=0.2)



    num_iterations = 1000
    loss_values = []
    for iteration in range(num_iterations):
        optimizer.zero_grad()
        transformed_volume = model(moving_volume)
        loss = ncc_loss(fixed_volume, transformed_volume)
        loss.backward(retain_graph=True)
        
        # Clip gradients
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        
        optimizer.step()
        # scheduler.step()
    
        loss_values.append(loss.item())


    transformed_volume = model(moving_volume)


    
    return model.tx.item(), model.ty.item(), transformed_volume, loss_values
```

```python
volume_corrected = np.copy(volume)
final_loss = []
for i in trange(volume.shape[2]-1):

    image1 = volume[:,:,i].astype(np.float32)
    
    image2 = volume[:,:,i+1].astype(np.float32)
    
    
    fixed_volume = torch_volume(image1)
    moving_volume = torch_volume(image2)
    
    tx, ty, transformed_volume, loss_values = register_2d(fixed_volume, moving_volume)
    volume_corrected[:,:,i+1] = transformed_volume.detach().numpy()

    plt.plot(loss_values)
    final_loss.append(loss_values)
```
